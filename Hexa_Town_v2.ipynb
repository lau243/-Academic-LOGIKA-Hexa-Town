```python
import random
import numpy as np

s=0
n=14
tax=2
p1=25
i1=10
p2=15
i2=15
r2=15
p3=35
p4=30
p5=15
i5=15
r5=15
p7=35

class player:
  def __init__(self, 
               name = '',
               score = 30, 
               loc = 1,
               hist = [],
               invt = [],
               acc = 0,
               dodge = 0):
    self.name = name
    self.score = score
    self.loc = loc
    self.hist = hist
    self.invt = invt
    self.acc = acc
    self.dodge = dodge

teamname = ['Alpha', 'Beta', 'Gamma', 'Delta', 'Epsilon']

alph = player(name = teamname[0], hist = [])
beta = player(name = teamname[1], hist = [])
gamm = player(name = teamname[2], hist = [])
delt = player(name = teamname[3], hist = [])
epsi = player(name = teamname[4], hist = [])

teamobj = [alph, beta, gamm, delt, epsi]

def numinput(que = 'Masukan angka: ', low = 0, high = 100):
  while True:
    result = input(que).split(',')
    authorize = input('konfirmasi input? (y) ')
    authorize = (authorize == 'y')
    try:
      result = [int(x) for x in result]
      check = [((x >= low) and (x <= high)) for x in result]
      authorize = authorize and (sum(check)==len(check))
    except:
      authorize = False
    if authorize:
      break
  if len(result) == 1:
    result = result[0]
  return result

def teaminput(que = 'Masukan nama tim : ', valid = teamname):
  global teamname
  while True:
    result = input(que)
    if result not in valid:
      print('invalid team')
    else:
      break
  i = teamname.index(result)
  return teamobj[i]

def textinput(que = 'Masukan text: ', valid = None):
  while True:
    result = input(que).split(',')
    authorize = input('konfirmasi input? (y) ')
    authorize = (authorize == 'y')
    if valid:
      check = [(x in valid) for x in result]
      authorize = authorize and (sum(check)==len(check))
    if authorize:
      break
  if len(result) == 1:
    result = result[0]
  return result 

def itemselection(que = 'Masukan pilihan item setiap tim: '):
  global teamobj
  while True:
    result = input(que).split(',')
    authorize = input('konfirmasi input? (y) ')
    check = [((x in teamobj[i].invt) or (x == '')) for i,x in enumerate(result)]
    authorize = (authorize == 'y') and (sum(check)==len(check))
    if authorize:
      break
  return result

auction = {'Roti Gerobak'   :'Semua peluang pada aktivitas distrik team meningkat 3% untuk setiap team di distrik yang dikunjungi',
           'Batu Idocrase'  :'Duplikasi satu item pada inventori team',
           'Bunga Raflesia' :'Pada sore hari, memaksa satu team di distrik yang dikunjungi untuk kehilangan 1 item atau 6 koin',
           'Congklak'       :'Memilih satu team di distrik yang dikunjungi, dan membagi rata seluruh koin dari team pengguna dan target'}

powerups=['Kopi Luwak','Kalender Jawa','Teh Jawa','Ondel-ondel','Lompat Karung','Mikrolet',
          'Merpati Pos','Wedang Ronde','Batu Bacan','Rawon', 'Roti Gerobak','Batu Idocrase',
          'Bunga Raflesia','Congklak','Batu Opal','Tahu Bulat','Gundu','Batik','Sate Padang',
          'Bakiak', 'Topeng Monyet','Egrang','Ketoprak','Ketupat','Gado-gado']

need_counter = {'Ondel-ondel' : 1, 'Teh Jawa': 1}

rute = {'D':[4,5],'E':[1,3,6],'F':[2,7],'A':[3,4],'B':[1,2,5],'C':[6,7],'G':[2,3],'H':[1,4,7],'I':[5,6]}

active = {}

def nextturn():
  global s, n, tax, active, teamobj, teamname, powerups, need_counter, rute

  if s == 0:
    input('---------------------\nSelamat Datang di FINAL MTC LOGIKA UI 2022\n---------------------\n')
    start=powerups[:10]
    print('Pilihan item starter\n',start[:5],'\n',start[5:],'\n')

    # random item 
    random.shuffle(start)
    for i,team in enumerate(teamobj):
      get = start[2*i: 2*i+2]
      team.invt = get
      print(f'Team {team.name} mendapatkan item starter: {get[0]}, {get[1]}')

  if s == 7:
    for i,team in teamobj:
      trace = set(team.hist)
      if len(trace) == 7:
        team.score += 6
        print(f'Team {team.name} mendapatkan bonus 6 koin untuk melewati semua distrik pada Hexa Town')

  s += 1

  if s > n:
    print('----------------------------\nLEADERBOARD FINAL MTC LOGIKA UI 2022 \n')
    scores = [team.score for team in teamobj]
    rank = [sorted(scores).index(x) for x in scores]
    print(rank)
    
    for i in range(5):
      team = teamname[rank.index(4-i)]
      print(f'Urutan ke-{i+1} : Team {team.name}')
    print('\nSelamat untuk pemenang MTC LOGIKA UI 2022 \nsampai ketemu di LOGIKA UI selanjutnya!')

  else:
    record = {team.name: team.score for team in teamobj}
    print('-----------')
    print(f'Hari ke-{s}')
        
    input('\npagi hari \nteam memilih soal dari panitia \n')
    hasil = numinput('pendapatan koin hasil mengerjakan soal (separator koma)\n\n', high=26)
    hasil = [int(x) for x in hasil]

    hasil = batik(hasil)

    for i,team in enumerate(teamobj):
      print(f'Team {team.name} mendapatkan gaji mengajar {hasil[i]}')
      team.score += hasil[i]

    input('\nPersiapan untuk berpindah tempat\nteam memberikan nama item yang akan digunakan hari ini ke LO\n')
    if 'Rawon' in active.keys():
      print(f'Item Rawon mempersilakan team {active["Rawon"].name} untuk menunda aktivasi setelah perpindahan tempat')
      
    print('siang hari \nteam berpindah tempat \n')

    gundu()

    input('')

    # berpindah tempat
    tujuan = numinput('tujuan perpindahan setiap tim (separator koma)\n\n', low=1, high=7)
    tujuan = np.array(tujuan)

    asal = np.array([team.loc for team in teamobj])

    cost = ((tujuan == 1) + (asal == 1) + (np.abs(tujuan-asal) == 1) + (np.abs(tujuan-asal) == 5))
    cost = cost.astype(bool)
    cost = 1 - cost.astype(int)

    for i,team in enumerate(teamobj):
      team.hist.append(team.loc)
      team.loc = tujuan[i]
      if (c:=cost[i]) == 1:
        print(f'Team {team.name} membayar 1 koin untuk berpindah')
      team.score -= c

    satepadang()
    
    print('\nitem tersedia untuk setiap tim')
    for team in teamobj:
      print(f'Inventori team {team.name}: {team.invt}')

    activation = itemselection('\naktivasi item setiap tim, maks 1 (separator koma)\n\n')
    for i,team in enumerate(teamobj):
      if (item:=activation[i]) != '':
        if item in need_counter.keys():
          active[item] = [team, need_counter[item]]
        else:
          active[item] = team
        team.invt.remove(item)

    batuidocrase()
    batubacan()
    mikrolet()
    lompatkarung()
    wedangronde()
    tahubulat(1)
    congklak()

    rotigerobak()

    tehjawa()
    batuopal()
    bungaraflesia()

    input('\nsore hari \npetugas pajak datang \n')
    rutechoice = merpatipos()
    if rutechoice is None:
      rutechoice = tahubulat(2)
    if rutechoice is None:
      rutechoice = random.choice(list(rute.keys()))
    
    print(f'petugas pajak melalui rute {rutechoice}')
    print(f'Distrik yang terkena pajak : {rute[rutechoice]}\n\n')

    ondelondel()
    currenttax = kopiluwak(tax)
    currenttax, act = kalenderjawa(currenttax)

    count = True
    for team in teamobj:
      if team.loc in rute[rutechoice]:
        if random.randint(1,100) <= team.dodge:
          print(f'Team {team.name} menghindari pajak')
        else:
          team.score -= currenttax
          print(f'Team {team.name} {act} {currenttax} koin')
          count = False
    if count:
      print('\ntidak ada team yang dipajak')
    
    bakiak(rutechoice)

    input('\nmalam hari \nrekap penghasilan semua team \n')
    for team in teamobj:
      new = team.score
      old = record[team.name]
      if new-old >= 0:
        diff = new - old
        sign = '+'
      else:
        diff = old - new
        sign = '-'
      print(f'Team {team.name}: {new} ({sign}{diff}), item: {team.invt}')
    
    #activation reset
    temp = {}
    for item,content in active.items():
      if (type(content) == list):
        if (content[1] > 0):
          temp[item] = [content[0], content[1]-1]
    active = temp
    
def satutempat(team):
  global teamobj
  result = [team2 for team2 in teamobj if team2.loc == team.loc]
  result.remove(team)
  return result

def batik(hasil):
  global active, teamobj, s
  if (item:='Batik') in active.keys():
    team = active[item]
    if s%7 not in [0,6]:
      i = teamobj.index(team)
      hasil[i] = 2*hasil[i]
      print(f'Item Batik menggandakan gaji mengajar team {team.name}')
    else:
      print(f'Item Batik team {team.name} hangus di hari Sabtu atau hari Minggu')
  return hasil

def gundu():
  global active, teamobj
  if (item:='Gundu') in active.keys():
    team = active[item]
    target = satutempat(team)

    print(f'Efek item Gundu oleh team {team.name}')
    for team2 in target:
      print(f'Item Gundu membuat team {team2.name} tidak bisa berpindah dari distrik {team.loc} dan kehilangan 1 koin')
      team2.score -=1

def satepadang():
  global active, teamobj
  if (item:='Sate Padang') in active.keys():
    team = active[item]
    target = satutempat(team)
  
    print(f'Efek item Sate Padang oleh team {team.name}')
    for team2 in target:
      print(f'Item Sate Padang membuat team {team2} tidak bisa mengaktifkan item pada giliran ini')
    

def batuidocrase():
  global active, teamobj
  if (item:='Batu Idocrase') in active.keys():
    team = active[item]
    if len(team.invt)>0:
      print(f'Item Batu Idocrase akan menduplikasi satu item dari team {team.name}\nitem pada inventori : {team.invt}')
      valid = [team.invt]
      selection = textinput('Item yang akan diduplikasi: ', valid = team.invt)
      team.invt.append(selection)
      print(f'Item {selection} berhasil diduplikasi')
    else:
      print('Item Batu Idocrase gagal menemukan target')

def batubacan():
  global active, teamobj
  if (item:='Batu Bacan') in active.keys():
    team = active[item]
    inc = 2 + (s-1)//3
    print(f'Item Batu Bacan memberikan {inc} koin kepada team {team.name}')
    team.score += inc

def mikrolet():
  global active, teamobj
  if (item:='Mikrolet') in active.keys():
    team = active[item]
    print(f'Item Mikrolet memberikan 5 koin kepada team {team.name} untuk modal berpindah distrik')
    team.score += 5

def lompatkarung():
  global active, teamobj
  if (item:='Lompat Karung') in active.keys():
    team = active[item]
    print('Item Lompat Karung memindahkan posisi peserta')
    print(f'Team {team.name} membayar 1 koin untuk mengaktifkan Lompat Karung, semua team akan berpindah ke seberang papan')
    team.score -= 1

    for team in teamobj:
      if team.loc == 1:
        print(f'Team {team.name} tidak berpindah')
      elif team.loc<5:
        print(f'Team {team.name} akan dipindahkan dari distrik {loc} ke distrik {loc+3}')
        team.loc +=3
      else:
        print(f'Team {team.name} akan dipindahkan dari distrik {loc} ke distrik {loc-3}')
        team.loc -=3
      
def wedangronde():
  global active, teamobj
  if (item:='Lompat Karung') in active.keys():
    team = active[item]
    if team.loc != 1:
      target = satutempat(team).append(team)

      print('Item Wedang Ronde memindahkan posisi peserta')
      print(f'Team {team.name} mengaktifkan Wedang Ronde, semua team yang akan pergi ke distrik {team.loc} pindah ke distrik 1')

      for team in target:
        print(f'Team {team.name} akan dipindahkan dari distrik {team.loc} ke distrik 1')
        team.loc = 1
        
def tahubulat(k):
  global active, teamobj, rute
  if (item:='Tahu Bulat') in active.keys():
    team = active[item]
    if k==1:
      print(f"""\nItem Tahu Bulat
                \nLO memberikan satu rute random (dari A sampai I) ke team {team.name}
                \nTeam dapat memilih untuk menggunakan rute pajak tersebut atau mengacak rute pajak""")
    elif k==2:
      print(f'Item Tahu Bulat mengizinkan team {team.name} untuk memilih rute pajak')
      rutechoice = textinput('Rute pajak yang diberikan LO: ', valid = rute.keys())
      confirm = textinput('Rute digunakan (y), atau Rute diacak (n)? :', valid = ['y','n'])
      if confirm == 'y':
        return rutechoice

def congklak():
  global active, teamobj
  if (item:='Congklak') in active.keys():
    team = active[item]
    target = satutempat(team)
    targetname = [team2.name for team2 in target]

    if target:
     print('Item Congklak akan memilih target\n\npilihan target:') 
     team2 = teaminput('Team yang akan menjadi target: ', valid = targetname)
     mid = (team.score + team2.score)//2
     team.score = mid
     team2.score = mid
     print(f'Skor team {team.name} dan {team2.name} menjadi rata-rata skor')
    else:
      print('Item Congklak tidak menemukan target')

def rotigerobak():
  global active, teamobj, p1, i1, p2, i2, r2, p3, p4, p5, i5, r5, p7
  if (item:='Roti Gerobak') in active.keys():
    team = active[item]
    loc = team.loc
    print(f'Team {team.name} mengaktifkan Roti Gerobak\nItem Roti Gerobak mengakibatkan peluang pada distrik {team.loc} meningkat 3%')
  else:
    loc = 0
  
  if loc == 1:
    Region1(p1+3, i1+3)
  else:
    Region1(p1, i1)
  if loc == 2:
    p2 = Region2(p2+3, i2+3, r2+3)
  else:
    p2 = Region2(p2, i2, r2)
  if loc == 3:
    Region3(p3+3)
  else:
    Region3(p3)
  if loc == 4:
    Region4(p4+3)
  else:
    Region4(p4)
  if loc == 5:
    p5 = Region5(p5+3, i5+3, r5+3)
  else:
    p5 = Region5(p5, i5, r5)
  Region6()
  if loc == 7:
    Region7(p7+3)
  else:
    Region7(p7)

def visitor(k):
  global teamobj
  result = [team for team in teamobj if team.loc == k]
  return result

def Region1(p1, i1):
  global teamobj
  for team in teamobj:
    team.dodge = 0
  if (visit:=visitor(1)):
    c = len(visit)
    input()
    input(f'Distrik 1 : Distrik Industri Besi\npartisipan = {c}\n')
    prob=p1+(c-1)*i1
    for team in visit:
      print(f'Team {team.name} mendapat peluang {prob}% menghindari pajak')
      team.dodge = prob
 
def Region2(p2, i2, r2):
  global teamobj
  if (visit:=visitor(2)):
    c = len(visit)
    input()
    input(f'\nDistrik 2 : Distrik Pertambangan \npartisipan = {c}\n')
    for team in visit:
      team.score -= 2
      print(f'Team {team.name} membayar 2 koin untuk menambang')
    if visit:
      input(f'\npeluang menambang dimulai dari {p2}%\nmari menambang...')
    while visit:
      team = teaminput('Team yang akan mencoba peruntungan: ', valid = [team.name for team in visit])
      visit.remove(team)
      if random.randint(1,100) <=p2:
        team.score += 12
        p2 = r2
        print(f'Team {team.name} berhasil menemukan mineral langka seharga 12 koin \npeluang mendapatkan mineral langka kembali ke {r2}%')
      else:
        p2 += i2
        print(f'Team {team.name} gagal mendapatkan mineral langka \npeluang mendapatkan mineral langka naik menjadi {p2}%')
  return p2

def Region3(p3):
  global teamobj, auction, powerups
  if (visit:=visitor(3)):
    c = len(visit)
    input()
    input(f'\nDistrik 3 : Distrik Makanan \npartisipan = {c} \n')
    for team in visit:
      input(f'Team {team.name} mencoba peruntungan mendapatkan random Item')
      if random.randint(1,100) <= p3:
        item = random.choice([x for x in powerups[14:] if x not in auction.keys()])
        print(f'Team {team.name} berhasil mendapatkan Item {item}')
        team.invt.append(item)
      else:
        print(f'Team {team.name} gagal mendapatkan Item')

def Region4(p4):
  global teamobj
  if (visit:=visitor(4)):
    c = len(visit)
    input()
    input(f'\nDistrik 4 : Distrik Administratif\npartisipan = {c}\n')
    for team in visit:
      nominal = numinput(f'Team {team.name} mendonasikan koin sebanyak (maksimal 15) : ', high = 15)
      if random.randint(1,100) <= p4:
        print(f'Team {team.name} mendapatkan untung sebesar {nominal}')
        team.score += nominal
      else:
        loss = nominal//2 + nominal%2
        print(f'Team {team.name} mendapatkan rugi sebesar {loss}')
        team.score -= loss

def Region5(p5, i5, r5):
  global teamobj, powerups
  if (visit:=visitor(5)):
    c = len(visit)
    input()
    input(f'\nDistrik 5 : Distrik Market\npartisipan = {c}\n')
    item = textinput('Powerup yang akan dijual : ', valid = powerups[14:])
    for team in visit:
      confirm = textinput(f'Apakah Team {team.name} ikut memesan item? (y/n) ', valid = ['y', 'n'])
      if confirm != 'y':
        visit.remove(team)
      else:
        print(f'Team {team.name} membayar 2 koin untuk memesan item')
        team.score -= 2
    if visit:
      print(f'\npeluang item dimulai dari {p5}%')
    while visit:
      team = teaminput('Team yang akan mencoba peruntungan: ', valid = [team.name for team in visit])
      visit.remove(team)
      if random.randint(1,100) <=p5 :
        team.invt.append(item)
        p5 = r5
        print(f'Team {team.name} berhasil mendapatkan powerup {item} \npeluang mendapatkan powerup kembali ke {r5}%')
      else:
        p5 += i5
        print(f'Team {team.name} gagal mendapatkan powerup \npeluang mendapatkan powerup naik menjadi {p5}%')
  return p5
    
def Region6():
  global active, teamobj, s, auction, powerups
  for team in teamobj:
    if (akun:=team.acc) < 20:
      akun += akun//4
    elif akun < 30:
      akun += akun//6-1
      if akun > 30:
        akun -= 1
    else:
      akun += akun//8-2
      if akun > 40:
        akun -= 1
    team.acc = akun
  if (visit:=visitor(6)):
    c = len(visit)
    input()  
    input(f'\nDistrik 6 : Distrik Perekonomian\npartisipan = {c}\n')
    if s%3 == 0:
      item = powerups[9+s//3]
      print(f'Selamat Datang di Pelelangan Item \n\nitem yang akan dilelang: {item} \neffect: {auction[item]}\n')
      for team in teamobj:
        print(f'Team {team.name}')
      print('\nHarga dibuka dari 3 koin')
      cost=3
      if c==1:
        bidder = visit[0]
      else:
        visitname = [team.name for team in visit]
        while True:
          team = teaminput('\nBidding oleh: ', valid = visitname)
          cost = numinput('Nominal bidding (maksmial 25): ', low = cost, high = 25)
          confirm = textinput('Ada lagi? (y/n) ', valid = ['y','n'])
          if confirm == 'n':
            bidder = team
            break
      print(f'\nSelamat, item {item} terjual pada {team.name} seharga {cost} koin')
      team.score -= cost
      team.invt.append(item)

    else:
      print('Layanan perbankan')
      for team in visit:
        print(f'\nSelamat datang pelanggan {team.name} di Layanan perbankan')
        layanan = numinput('Ketik 1 untuk menabung, ketik 2 untuk menarik tabungan, \nketik 3 untuk tidak melakukan transaksi \n',
                           low = 1, high = 3)
        if layanan == 1:
          if team.acc > 0:
            print(f'Team {team.name} sudah memiliki akun aktif di Distrik Perekonomian')
          else:
            deposit = numinput('Masukan jumlah koin yang ingin ditabung (maksimal 10): ', high = 10)
            team.score -= deposit
            team.acc = deposit
            print('Terima kasih untuk kepercayaan anda menabung di Distrik Perekonomian')
        elif layanan == 2:
          nominal = team.acc
          team.score += nominal
          team.acc = 0
          print(f'Terima kasih, tabungan anda sudah selesai ditarik\nTabungan yang telah dicarikan sebesar {nominal} koin')

def Region7(p7):
  global active, teamobj
  if (visit:=visitor(7)):
    c = len(visit)
    input()
    input(f'\nDistrik 7 : Distrik Pariwisata\npartisipan = {c}\n')
    for team in visit:
      input(f'\nTeam {team.name} mencoba peruntungan mendapatkan 5 koin')
      if random.randint(1,100) <= p7:
        team.score += 5
        print(f'Team {team.name} berhasil mendapatkan 5 koin')
      else: 
        print(f'Team {team.name} gagal mendapatkan koin')

def tehjawa():
  global active, teamobj
  if (item:='Teh Jawa') in active.keys():
    team = active[item][0]
    
    target = satutempat(team)
    print(f'\nTeam {team.name} mengaktifkan Teh Jawa\nSemua team yang berada di distrik {team.loc} akan membeli teh seharga 1 koin kepada team {team.name}')
    for team2 in target:
      team2.score -= 1
      team.score += 1
      print(f'Team {team2.name} membayar 1 koin')

def batuopal():
  global active, teamobj
  if (item:='Batu Opal') in active.keys():
    team = active[item]  
    target = satutempat(team)
    targetname = [team2.name for team2 in target]

    print(f'\nTeam {team.name} mengaktifkan Batu Opal')
    if target:
      print(f'Item Batu Opal akan memilih target\n\npilihan target:\n')
      for team2 in target:
        print(f'{team2.name} dengan item: {team2.invt}')
      team2 = teaminput('Team yang akan menjadi target: ', valid = targetname)
      print(f'Item milik target:\n {team2.invt}\n')
      item = textinput('Item yang akan diduplikasi: ', valid = team2.invt)
      team.invt.append(item)
      print(f'Item {item} berhasil diduplikasi')
    else:
      print('Item Batu Opal gagal menemukan target')

def bungaraflesia():
  global active, teamobj
  if (item:='Bunga Raflesia') in active.keys():
    team = active[item]
    target = satutempat(team)
    targetname = [team2.name for team2 in target]

    print(f'\nTeam {team.name} mengaktifkan Bunga Raflesia')
    if target:
      print(f'Item Bunga Raflesia akan memilih target\n\npilihan target:\n{targetname}')
      team2 = teaminput('Team yang akan menjadi target: ', valid = targetname)
      print(f'Item milik target:\n {team2.invt}\n')

      choice = textinput('Target memilih kehilangan (koin atau nama item): ', valid = team2.invt + ['koin'])
      if choice == 'koin':
        team2.score -= 6
        print('Target kehilangan 6 koin')
      else:
        team2.invt.remove(choice)
        print(f'Target kehilangan item {choice}')
    else:
      print('Item Bunga Raflesia gagal menemukan target')

def merpatipos():
  global active, teamobj, rute
  if (item:='Merpati Pos') in active.keys():
    team = active[item]
    rutechoice = textinput(f'Item Merpati Pos\nTeam {team.name} memilih rute pajak: ', valid = rute.keys())
    return rutechoice

def ondelondel():
  global active, teamobj
  if (item:='Ondel-ondel') in active.keys():
    team = active[item][0]
    team.dodge = 101
    print(f'Item Ondel-ondel memberikan dodge 100% kepada team {team.name}')

def kopiluwak(tax):
  global active
  if (item:='Kopi Luwak') in active.keys():
    print('Item Kopi Luwak menyebabkan nominal pajak dua kali lipat')
    return 2*tax
  else:
    return tax

def kalenderjawa(tax):
  global active
  if (item:='Kalender Jawa') in active.keys():
    print('\nItem Kalender Jawa menyebabkan petugas pajak membagikan cukai')
    return -tax, 'mendapatkan'
  else:
    return tax, 'dipajak'

def bakiak(rutechoice):
  global active, teamobj, rute
  if (item:='Bakiak') in active.keys():
    team = active[item]
    if team.loc in rute[rutechoice]:
      team.score -= 2
      print(f'Item Bakiak membuat team {team.name} kehilangan 4 koin termasuk pajak')
    else:
      team.score += 7
      print(f'Item Bakiak membuat team {team.name} mendapatkan 5 koin karena memenangkan lomba bakiak')
  
```
