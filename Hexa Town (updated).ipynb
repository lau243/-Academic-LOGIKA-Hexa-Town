```python
import math, random
 
class HexaTown():
  def __init__(self,s=0,n=14,tax=2,mode='Normal',p1=25,i1=10,p2=15,i2=15,r2=15,p3=35,p4=30,p5=15,i5=15,r5=15,p7=35,
               teams=['Alpha','Beta','Gamma','Delta','Epsilon'],
               skor={'Alpha':30,'Beta':30,'Gamma':30,'Delta':30,'Epsilon':30},
               dest={'Alpha':0,'Beta':0,'Gamma':0,'Delta':0,'Epsilon':0},
               history={'Alpha':[],'Beta':[],'Gamma':[],'Delta':[],'Epsilon':[]},
               items={'Alpha':[],'Beta':[],'Gamma':[],'Delta':[],'Epsilon':[]},
               active={'Alpha':None,'Beta':None,'Gamma':None,'Delta':None,'Epsilon':None},
               account={'Alpha':[0,0],'Beta':[0,0],'Gamma':[0,0],'Delta':[0,0],'Epsilon':[0,0]},
               auct=['Batu Idocrase','Bunga Raflesia','Roti Gerobak','Congklak'],
               auct_eff={'Roti Gerobak':'Semua peluang pada aktivitas distrik team meningkat 3% untuk setiap team di distrik yang dikunjungi',
                     'Batu Idocrase':'Duplikasi satu item pada inventori team',
                     'Bunga Raflesia':'Pada sore hari, memaksa satu team di distrik yang dikunjungi untuk kehilangan 1 item atau 6 koin',
                     'Congklak':'Memilih satu team di distrik yang dikunjungi, dan membagi rata seluruh koin dari team pengguna dan target'},
               powerups=['Kopi Luwak','Kalender Jawa','Teh Jawa','Ondel-ondel','Lompat Karung','Mikrolet','Merpati Pos','Wedang Ronde','Batu Bacan','Rawon',
                         'Roti Gerobak','Batu Idocrase','Bunga Raflesia','Congklak','Batu Opal','Tahu Bulat','Gundu','Batik','Sate Padang','Bakiak',
                         'Topeng Monyet','Egrang','Ketoprak','Ketupat','Gado-gado'],
               rute={'D':[4,5],'E':[1,3,6],'F':[2,7],'A':[3,4],'B':[1,2,5],'C':[6,7],'G':[2,3],'H':[1,4,7],'I':[5,6]}):
    self.s=s
    self.n=n
    self.tax=tax
    self.mode=mode
    self.p1=p1
    self.i1=i1
    self.p2=p2
    self.i2=i2
    self.r2=r2
    self.p3=p3
    self.p4=p4
    self.p5=p5
    self.i5=i5
    self.r5=r5
    self.p7=p7
    self.teams=teams
    self.skor=skor
    self.dest=dest
    self.history=history
    self.items=items
    self.active=active
    self.account=account
    self.auct=auct
    self.auct_eff=auct_eff
    self.powerups=powerups
    self.rute=rute

    if self.mode !='Normal':
      for team in self.teams:
        self.items[team]=powerups.copy()
    
  def NextTurn(self):
    if self.s<self.n:
      if self.s==0:
        input('---------------------\nSelamat Datang di FINAL MTC LOGIKA UI 2022\n---------------------\n')
        start=self.powerups[0:10]
        print('Pilihan item starter\n',start)
        print()
        for team in self.teams:
          item=random.choice(start)
          print('Team {0} mendapatkan item starter pertama {1}'.format(team,item))
          self.items[team].append(item)
          start.remove(item)
        print()
        for i in range(5):
          team=self.teams[4-i]
          item=random.choice(start)
          print('Team {0} mendapatkan item starter kedua {1}'.format(team,item))
          self.items[team].append(item)
          start.remove(item)
  
      elif self.s==7:
        for team in self.teams:
          count=True
          for i in range(1,8):
            if self.history[team].count(i)==0:
              count=False
          if count:
            print('Team {0} mendapatkan bonus 6 koin untuk melewati semua distrik pada Hexa Town'.format(team))
            self.skor[team]+=6

      form=self.skor.copy()
      self.s+=1
      print('-----------')
      print('Hari ke-{0}'.format(self.s))
      
      input('\npagi hari \nteam memilih soal dari panitia \n')
      print('pendapatan koin hasil mengerjakan soal')
      temp={}
      for t in self.teams:
        while True:
          try:
            nilai=int(input('Team {0} mendapatkan : '.format(t)))
            if nilai>26:
              print('Nominal tidak valid :(')
            else:
              break
          except:
            print('Nominal tidak valid :(')  
        temp[t]=nilai
        
      while True:
        que=input('\nAda yang salah input? ')
        if que=='tidak' or que=='tdk' or que=='no':
          break
        else:
          try:
            team=input('Input nama team: ')
            nilai=int(input('Input nominal '))
            if team in self.teams:
              temp[team]=nilai
            else:
              print('Input tidak valid :(')
          except:
            print('Input tidak valid :(')

      for team in self.teams:
        if self.active[team]=='Batik':
          if (self.s-1)%7!=5 or (self.s-1)%7!=6:
            temp[team]=2*temp[team]
            print('Item Batik menggandakan gaji mengajar team {0}'.format(team))
          else:
            print('Item Batik hangus karena digunakan di hari Sabtu atau hari Minggu')
          self.skot[team]+=temp[team]
      
      input('\nPersiapan untuk berpindah tempat\nteam memberikan nama item yang akan digunakan hari ini ke LO\n')
      for team in self.teams:
        if self.active[team]=='Rawon':
          print('Item Rawon mempersilakan team {0} untuk menunda aktivasi setelah perpindahan tempat'.format(team))
      
      input('siang hari \nteam berpindah tempat \n')

      lis=[]
      if 'Sate Padang' in self.active.values():
        for team in self.teams:
          if self.active[team]=='Sate Padang':
            for t in self.teams:
              if self.dest[team]==self.dest[t] and t!=team:
                lis.append(t)

      if 'Gundu' in self.active.values():
        for team in self.teams:
          if self.active[team]=='Gundu':
            gun=self.dest[team]
            owner=team
      else:
        gun=0
      
      temp={}
      for t in self.teams:
        prev=self.dest[t]
        if gun>0 and prev==gun and owner!=t:
          lokasi=prev
          self.skor[t]-=1
          print('Item Gundu membuat team {0} tidak bisa berpindah dan kehilangan 1 koin'.format(t))
        else:
          while True:
            try:
              lokasi=int(input('Team {0} menuju distrik nomor : '.format(t)))
              if lokasi in range(1,8):
                break
              else:
                print('Nomor distrik tidak valid :(')
            except:
              print('Nomor distrik tidak valid :(')
        if lokasi!=1 and self.s>1 and prev>1 and abs(lokasi-prev)<5 and abs(lokasi-prev)>1:
          print('Team {0} membayar 1 koin untuk berpindah'.format(t))
          self.skor[t]-=1
        temp[t]=lokasi
      while True:
        que=input('\nAda yang salah input? ')
        if que=='tidak' or que=='tdk' or que=='no':
          break
        else:
          try:
            team=input('Input nama team: ')
            nilai=int(input('Input nomor distrik '))
            if team in self.teams:
              temp[team]=nilai
            else:
              print('Input tidak valid :(')
          except:
            print('Input tidak valid :(')
      for team in self.teams:
        self.history[team].append(temp[team])
        self.dest[team]=temp[team]
      
      print('\nAktivasi Item')
      temp={}
      for team in self.teams:
        if self.active[team]=='Ondel':
          print('Item yang digunakan team {0} pada hari ini: Ondel-ondel'.format(team))
        elif self.active[team]=='Teh Jawa1':
          print('Item yang digunakan team {0} pada hari ini: Teh Jawa'.format(team))
        else:
          self.active[team]= None 
          if team in lis:
            print('Item Sate Padang membuat team {0} tidak bisa mengaktifkan item pada giliran ini'.format(team))
          else:
            while True:
              item=input('Item yang digunakan team {0} pada hari ini: '.format(team))
              if item=='' or item=='tidak' or item=='no' or item=='tdk':
                break
              if item not in self.powerups:
                print('item tidak valid\n')
              else:
                self.active[team]=item
                self.items[team].remove(item)
                break
      while True:
        que=input('\nAda yang salah input? ')
        if que=='tidak' or que=='tdk' or que=='no':
          break
        else:
          try:
            team=input('Input nama team: ')
            item=input('Input nama item ')
            if item in self.powerups and team in self.teams:
              temp[team]=item
            else:
              print('Input tidak valid :(')
          except:
            print('Input tidak valid :(')
      for team in self.teams:
        try:
          self.active[team]=temp[team]
        except:
          self.active[team]=None

      if 'Batu Idocrase' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Batu Idocrase':
            print('Item Batu Idocrase akan menduplikasi satu item dari team {0}'.format(team))
            print('item pada inventori :',self.items[team])
            if self.items[team]!=[]:
              while True:
                item=input('Item yang akan diduplikasi: ')
                if item in self.items[team]:
                  self.items[team].append(item)
                  print('Item {0} berhasil diduplikasi'.format(item))
                  break
                else:
                  print('Item tidak ada pada inventori')
            else:
              print('Item Batu Idocrase gagal menemukan target')

      if 'Batu Bacan' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Batu Bacan':
            inc=2+math.floor((self.s-1)/3)
            print('Item Batu Bacan memberikan {0} koin kepada team {1}'.format(inc,team))
            self.skor[team]+=inc

      if 'Mikrolet' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Mikrolet':
            print('Item Mikrolet memberikan 5 koin kepada team {0} untuk modal berpindah distrik'.format(team))
            self.skor[team]+=5

      if 'Lompat Karung' in self.active.values():
        print()
        print('Item Lompat Karung memindahkan posisi peserta')
        for team in self.teams:
          if self.active[team]=='Lompat Karung':
            print('Team {0} membayar 1 koin untuk mengaktifkan Lompat Karung, semua team akan berpindah ke seberang papan'.format(team))
            self.skor[team]-=1
        print()
        for team in self.teams:
          l=self.dest[team]
          if l==1:
            print('Team {0} tidak berpindah'.format(team))
          elif l<5:
            print('Team {0} akan dipindahkan dari distrik {1} ke distrik {2}'.format(team,l,l+3))
            self.dest[team]=l+3
          else:
            print('Team {0} akan dipindahkan dari distrik {1} ke distrik {2}'.format(team,l,l-3))
            self.dest[team]=l-3
      
      if 'Wedang Ronde' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Wedang Ronde' and self.dest[team]!=1:
            print('Item Wedang Ronde memindahkan posisi peserta')
            print('Team {0} mengaktifkan Wedang Ronde, semua team yang akan pergi ke distrik {1} pindah ke distrik 1'.format(team,self.dest[team]))
            for t in self.teams:
              if self.dest[t]==self.dest[team] and t!=team:
                print('Team {0} akan dipindahkan dari distrik {1} ke distrik 1'.format(t,self.dest[t]))
                self.dest[t]=1

      if 'Tahu Bulat' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Tahu Bulat':
            print('\nItem Tahu Bulat\nLO memberikan satu rute random (dari A sampai I) ke team {0}\nTeam dapat memilih untuk menggunakan rute pajak tersebut atau mengacak rute pajak'.format(team))
      
      if 'Congklak' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Congklak':
            print('\nTeam {0} mengaktifkan Congklak'.format(team))
            c= list(self.dest.values()).count(self.dest[team])
            if c==1:
              print('Item Congklak tidak menemukan target')
            else:
              print('Item Congklak akan memilih target\n\npilihan target:')
              for t in self.teams:
                if self.dest[t]==self.dest[team] and t!=team:
                  print(t,', dengan skor ',self.skor[t])
              while True:
                target=input('Team yang akan menjadi target: ')
                if target not in self.teams or target==team or self.dest[target]!=self.dest[team]:
                  print('Target tidak valid')
                else:
                  med=math.floor((self.skor[target]+self.skor[team])/2)
                  self.skor[target]=med
                  self.skor[team]=med
                  print('Skor team {0} dan {1} menjadi rata-rata skor'.format(team,target))
                  break

      
      dodge=self.RegionEffect()
     
      if 'Teh Jawa' in self.active.values() or 'Teh Jawa1' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Teh Jawa' or self.active[team]=='Teh Jawa1':
            print('\nTeam {0} mengaktifkan Teh Jawa'.format(team))
            lok=self.dest[team]
            print('Semua team yang berada di distrik {0} akan membeli teh seharga 1 koin kepada team {1}'.format(lok,team))
            for t in self.teams:
              if self.dest[t]==self.dest[team] and t!=team:
                print('Team {0} membayar 1 koin'.format(t))
                self.skor[t]-=1
                self.skor[team]+=1
          if self.active[team]=='Teh Jawa':
            self.active[team]='Teh Jawa1'
          elif self.active[team]=='Teh Jawa1':
            self.active[team]=None

      if 'Batu Opal' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Batu Opal':
            print('\nTeam {0} mengaktifkan Batu Opal'.format(team))
            c= list(self.dest.values()).count(self.dest[team])
            if c==1:
              print('Item Batu Opal tidak menemukan target')
            else:
              print('Item Batu Opal akan memilih target\n\npilihan target:')
              for t in self.teams:
                if self.dest[t]==self.dest[team] and t!=team:
                  print(t,' dengan item: ',self.items[t])
              while True:
                target=input('Team yang akan menjadi target: ')
                if target not in self.teams or target==team or self.dest[target]!=self.dest[team]:
                  print('Target tidak valid')
                else:
                  break
              print('Item milik target:\n',self.items[target])
              if self.items[target]!=[]:
                while True:
                  item=input('Item yang akan diduplikasi: ')
                  if item in self.items[target]:
                    self.items[team].append(item)
                    print('Item {0} berhasil diduplikasi'.format(item))
                    break
                  else:
                    print('Item tidak ada pada inventori')
              else:
                print('Item Batu Opal gagal menemukan target item')

      if 'Bunga Raflesia' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Bunga Raflesia':
            print('\nTeam {0} mengaktifkan Bunga Raflesia'.format(team))
            c= list(self.dest.values()).count(self.dest[team])
            if c==1:
              print('Item Bunga Raflesia tidak menemukan target')
            else:
              print('Item Bunga Raflesia akan memilih target\n\npilihan target:')
              for t in self.teams:
                if self.dest[t]==self.dest[team] and t!=team:
                  print(t)
              while True:
                target=input('Team yang akan menjadi target: ')
                if target not in self.teams or target==team or self.dest[target]!=self.dest[team]:
                  print('Target tidak valid')
                else:
                  break
              print('Item milik target:\n',self.items[target])
              while True:
                x=input('Target memilih kehilangan (koin atau nama item): ')
                if x=='koin':
                  self.skor[target]-=6
                  print('Target kehilangan 6 koin')
                  break
                elif x in self.items[target]:
                  self.items[target].remove(x)
                  print('Target kehilangan {0}'.format(x))
                  break
                else:
                  print('Input tidak valid')        
      
      input('\nsore hari \npetugas pajak datang \n')
      if 'Merpati Pos' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Merpati Pos':
            while True:
              try:
                x=input('Team {0} memilih rute pajak: '.format(team))
                if x in self.rute.keys():
                  break
                else:
                  print('Rute tidak valid :(')
              except:
                print('Rute tidak valid :(')
      elif 'Tahu Bulat' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Tahu Bulat':
            print('Item Tahu Bulat mengizinkan team {0} untuk memilih rute pajak'.format(team))
            while True:
              dec=input('Apakah Rute akan dipakai, atau diacak? ')
              if dec=='dipakai' or dec=='pakai':
                x=input('Rute pajak yang diberikan LO: ')
                break
              elif dec=='diacak' or dec=='acak':
                x=random.choice(list(self.rute.keys()))
                break
              else:
                print('Perintah tidak valid')
      else:
        x=random.choice(list(self.rute.keys()))
      print('petugas pajak melalui rute {0}'.format(x))
      print('Distrik yang terkena pajak : ')
      for i in self.rute[x]:
        print(i)
      print('')
      k=0

      for team in self.teams:
        if self.active[team]=='Ondel-ondel' or self.active[team]=='Ondel':
          dodge[team]=100
          print('Item Ondel-ondel memberikan dodge 100% kepada team {0}'.format(team))
          if self.active[team]=='Ondel-ondel':
            self.active[team]='Ondel'
          elif self.active[team]=='Ondel':
            self.active[team]= None
      
      tax=self.tax
      if 'Kopi Luwak' in self.active.values() and 'Kalender Jawa' in self.active.values():
        print('\nItem Kopi Luwak dan Kalender Jawa menyebabkan petugas pajak membagikan dua kali cukai')
        tax=-2*tax
      elif 'Kopi Luwak' in self.active.values():
        print('\nItem Kopi Luwak menyebabkan petugas pajak menarik nominal pajak dua kali lebih besar')
        tax=2*tax
      elif 'Kalender Jawa' in self.active.values():
        print('\nItem Kalender Jawa menyebabkan petugas pajak membagikan cukai')
        tax=-tax

      k=True
      print()
      for t in self.teams:
        if self.dest[t] in self.rute[x]:
          k=False
          y=random.randint(1,100)
          if y>dodge[t]:
            self.skor[t]-=tax
            if 'Kalender Jawa' in self.active.values():
              print('Team {0} mendapatkan {1} koin'.format(t,abs(tax)))
            else:
              print('Team {0} dikenai pajak {1} koin'.format(t,tax))
          else:
            print('Team {0} behasil menghindari pajak'.format(t))
      if k:
        print('tidak ada team yang dikenai pajak')
      
      if 'Bakiak' in self.active.values():
        print()
        for team in self.teams:
          if self.active[team]=='Bakiak':
            if self.dest[team] in self.rute[x]:
              self.skor[team]-=2
              print('Item Bakiak membuat team {0} kehilangan 2 koin karena kena pajak'.format(team))
            else:
              self.skor[team]+=5
              print('Item Bakiak membuat team {0} mendapatkan 5 koin karena memenangkan lomba bakiak'.format(team))

      input('\nmalam hari \nrekap penghasilan semua team \n')
      for t in self.teams:
        if self.skor[t]-form[t]>=0:
          sign='+'
        else:
          sign=''
        print('Team {0} : {1} ({2}{3}), item={4}'.format(t,self.skor[t],sign,self.skor[t]-form[t],self.items[t]))
      input()
    if self.s==self.n:
      self.s+=1
      print('\ncorrection term by game master (if needed)')
      for t in self.teams:
        while True:
          try:
            nilai=int(input('Team {0} correction term : '.format(t)))
          except:
            print('Nominal tidak valid :(')
          else:
            break  
        self.skor[t]+=nilai

      print('\nLEADERBOARD FINAL MTC LOGIKA UI 2022 \n')
      order=sorted(list(set(self.skor.values())),reverse=True)
      for i in range(len(order)):
        for t in self.teams:
          if self.skor[t]==order[i]:
            print('Urutan ke-{0} : Team {1}'.format(i+1,t))
      print('\nSelamat untuk pemenang MTC LOGIKA UI 2022 \nsampai ketemu di LOGIKA UI selanjutnya!')
 
  def RegionEffect(self):
    if 'Roti Gerobak' in self.active.values():
      print()
      for team in self.teams:
        if self.active[team]=='Roti Gerobak':
          print('Team {0} mengaktifkan Roti Gerobak\nItem Roti Gerobak mengakibatkan peluang pada distrik {1} meningkat 3%'.format(team,self.dest[team]))
          if self.dest[team]==1:
            self.p1+=3
            self.i1+=3
          elif self.dest[team]==2:
            self.p2+=3
            self.i2+=3
            self.r2+=3
          elif self.dest[team]==3:
            self.p3+=3
          elif self.dest[team]==4:
            self.p4+=3
          elif self.dest[team]==6:
            self.p5+=3
            self.i5+=3
            self.r5+=3
          elif self.dest[team]==7:
            self.p7+=3
    dodge=self.Region1()
    self.Region2()
    self.Region3()
    self.Region4()
    self.Region5()
    self.Region6()
    self.Region7()
    if 'Roti Gerobak' in self.active.values():
      for team in self.teams:
        if self.active[team]=='Roti Gerobak':
          if self.dest[team]==1:
            self.p1-=3
            self.i1-=3
          elif self.dest[team]==2:
            self.i2-=3
            self.r2-=3
          elif self.dest[team]==3:
            self.p3-=3
          elif self.dest[team]==4:
            self.p4-=3
          elif self.dest[team]==6:
            self.i5-=3
            self.r5-=3
          elif self.dest[team]==7:
            self.p7-=3
    return dodge

  def Region1(self):
    p1=self.p1
    i1=self.i1
    c= list(self.dest.values()).count(1)
    dodge={'Alpha':0,'Beta':0,'Gamma':0,'Delta':0,'Epsilon':0}
    if c>0:
      input()
      input('\nDistrik 1 : Distrik Industri Besi\npartisipan = {0}\n'.format(c))
      for t in self.teams:
        if self.dest[t]==1:
          prob=p1+(c-1)*i1
          print('Team {0} mendapat peluang {1}% menghindari pajak'.format(t,prob))
          dodge[t]=prob
    return dodge
    
  def Region2(self):
    p2=self.p2
    i2=self.i2
    r2=self.r2
    c= list(self.dest.values()).count(2)
    if c>0:
      input()
      input('\nDistrik 2 : Distrik Pertambangan \npartisipan = {0}\n'.format(c))
      att=[]
      for t in self.teams:
        if self.dest[t]==2:
          att.append(t)
          self.skor[t]-=2
          print('Team {0} membayar 2 koin untuk menambang'.format(t))
      input('\npeluang menambang dimulai dari {0}%\nmari menambang...'.format(p2))
      for i in range(len(att)):
        while True:
          t=input('Team yang akan mencoba peruntungan : ')
          if t in att:
            break
          else:
            print('Team tidak hadir di Distrik Pertambangan')
        x=random.randint(1,100)
        if x<=p2:
          self.skor[t]+=12
          p2=r2
          print('Team {0} berhasil menemukan mineral langka seharga 12 koin \npeluang mendapatkan mineral langka kembali ke {1}%'.format(t,r2))
        else:
          p2+=i2
          print('Team {0} gagal mendapatkan mineral langka \npeluang mendapatkan mineral langka naik menjadi {1}%'.format(t,p2))
      self.p2=p2
      
  def Region3(self):
    p3=self.p3
    c= list(self.dest.values()).count(3)
    if c>0:
      input()
      input('\nDistrik 3 : Distrik Makanan \npartisipan = {0} \n'.format(c))
      for t in self.teams:
        if self.dest[t]==3:
          input('Team {0} mencoba peruntungan mendapatkan random Item'.format(t))
          x=random.randint(1,100)
          if x<=p3:
            while True:
              y=random.choice(self.powerups[14:])
              if y not in self.auct:
                break
            self.items[t].append(y)
            print('Team {0} berhasil mendapatkan Item {1}'.format(t,y))
          else:
            print('Team {0} gagal mendapatkan Item'.format(t))
    
  def Region4(self):
    p4=self.p4
    c= list(self.dest.values()).count(4)
    if c>0:
      input()
      input('\nDistrik 4 : Distrik Administratif\npartisipan = {0}\n'.format(c))
      for t in self.teams:
        if self.dest[t]==4:
          while True:
            try:
              y=int(input('Team {0} mendonasikan koin sebanyak (maksimal 15) : '.format(t)))
            except:
              print('Nominal tidak valid :(')
            else:
              break
          x=random.randint(1,100)
          if x<=p4:
            self.skor[t]+=y
            print('Team {0} mendapatkan untung sebesar {1}'.format(t,y))
          else:
            self.skor[t]-=math.ceil(y/2)
            print('Team {0} mendapatkan rugi sebesar {1}'.format(t,math.ceil(y/2)))
    
  def Region5(self):
    p5=self.p5
    i5=self.i5
    r5=self.r5
    c= list(self.dest.values()).count(5)
    if c>0:
      input()
      input('\nDistrik 5 : Distrik Market\npartisipan = {0}\n'.format(c))
      while True:
        item=input('Powerup yang akan dijual : ')
        if item in self.powerups[14:]:
          break
      print('')
      att=[]
      for t in self.teams:
        if self.dest[t]==5:
          y=input('Apakah Team {0} ikut memesan item? '.format(t))
          if y=='iya' or y=='yes' or y=='ya':
            self.skor[t]-=2
            print('Team {0} membayar 2 koin itun memesan item'.format(t))
            att.append(t)
      if len(att)>0:
        print('\npeluang item dimulai dari {0}%'.format(p5))
      for i in range(len(att)):
        while True:
          t=input('Team yang akan mencoba peruntungan : ')
          if t in att:
            break
          else:
            print('Team tidak hadir di Distrik Market')
        x=random.randint(1,100)
        if x<=p5:
          self.items[t].append(item)
          p5=r5
          print('Team {0} berhasil mendapatkan powerup {1} \npeluang mendapatkan powerup kembali ke {2}%'.format(t,item,r5))
        else:
          p5+=i5
          print('Team {0} gagal mendapatkan powerup \npeluang mendapatkan powerup naik menjadi {1}%'.format(t,p5))
      self.p5=p5
    
  def Region6(self):
    c= list(self.dest.values()).count(6)
    for team in self.teams:
      akun=self.account[team][0]
      mark=self.account[team][1]
      if akun<20:
        akun+=math.floor(akun/4)
      elif akun<30:
        akun+=math.floor(akun/6)
        if akun>30:
          akun-=2
        else:
          akun-=1
      else:
        akun+=math.floor(akun/8)
        if akun>40:
          akun-=3
        else:
          akun-=2
      self.account[team][0]=akun

    if c>0:
      input()
      input('\nDistrik 6 : Distrik Perekonomian\npartisipan = {0}\n'.format(c))
      if self.s%3==0:
        it=self.auct[int(self.s/3)-1]
        print('Selamat Datang di Pelelangan Item \n\nitem yang akan dilelang: {0} \neffect: {1}\n'.format(it,self.auct_eff[it]))
        print('Harga dibuka dari 3 koin')
        cost=3
        if c==1:
          for team in self.teams:
            if self.dest[team]==6:
              bidder=team
        while True:
          print()
          try:
            bidder=input('Bidding oleh : ')
            if self.dest[bidder]==6:
              while True:
                try:
                  cost1=int(input('Nominal bidding (maksimal 25): '))
                except:
                  print('Nominal tidak valid')
                else:
                  break
              if cost1<=cost:
                print('Bid tidak cukup tinggi')
              else:
                cost=cost1
                cont=input('Ada lagi? ketik (tidak) untuk menghentikan pelelangan ')
                if cont=='tidak' or cont=='tdk' or cont=='no' or cost==25:
                  break
          except:
            print("Bidder tidak hadir di Distrik Perekonomian")
        
        print()
        print('Selamat, item {0} terjual pada {1} seharga {2} koin'.format(it,bidder,cost))
        self.items[bidder].append(it)
        self.skor[bidder]-=cost
      else:
        print('Layanan perbankan')
        for team in self.teams:
          if self.dest[team]==6:
            print()
            print('Selamat datang pelanggan {0} di Layanan perbankan'.format(team))
            while True:
              try:
                lay=int(input('Ketik 1 untuk menabung, ketik 2 untuk menarik tabungan, \nketik 3 untuk tidak melakukan transaksi \n'))
                if lay in range(1,4):
                  break
                else:
                  print('Jenis layanan tidak valid, silakan mencoba lagi')
              except:
                print('Jenis layanan tidak valid, silakan mencoba lagi')
            if lay==1:
              if self.account[team][1]>0:
                print('Team {0} sudah pernah menabung di Distrik Perekonomian')
              else:
                while True:
                  try:
                    deposit=int(input('Masukan jumlah koin yang ingin ditabung (maksimal 10): '))
                    if deposit in range(11):
                      break
                    else:
                      print('Nominal tidak valid')
                  except:
                    print('Nominal tidak valid')
                self.skor[team]-=deposit
                self.account[team][0]+=deposit
                self.account[team][1]+=1
                print('Terima kasih untuk kepercayaan anda menabung di Distrik Perekonomian')
            elif lay==2:
              nom=self.account[team][0]
              print('Terima kasih, tabungan anda sudah selesai ditarik')
              print('Tabungan yang telah dicarikan sebesar {0} koin'.format(nom))
              self.skor[team]+=nom
              self.account[team][0]=0
              self.account[team][1]=0      

  def Region7(self):
    p7=self.p7
    c= list(self.dest.values()).count(7)
    if c>0:
      input()
      input('\nDistrik 7 : Distrik Pariwisata\npartisipan = {0}\n'.format(c))
      for t in self.teams:
        if self.dest[t]==7:
          input('\nTeam {0} mencoba peruntungan mendapatkan 5 koin'.format(t))
          x=random.randint(1,100)
          if x<=p7:
            self.skor[t]+=5
            print('Team {0} berhasil mendapatkan 5 koin'.format(t))
          else: 
            print('Team {0} gagal mendapatkan koin'.format(t))
```
